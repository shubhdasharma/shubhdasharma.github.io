### Plotting histograms of all the bin channels for each image

```python
fig, ax = plt.subplots(5,1,figsize=(15,15)) # 3 color channels and 1 image plot

ax[0].hist(im_data1[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(im_data1[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(im_data1[:,:,2].flatten(), bins=256, color='blue') # Blue Channel
ax[3].hist(im_data1[:,:,3].flatten(), bins=256, color='orange') # Alpha Channel

ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')
ax[0].set_ylabel('Number of pixels')
ax[1].set_ylabel('Number of pixels')
ax[2].set_ylabel('Number of pixels')
ax[3].set_ylabel('Number of pixels')

# finally, include my image:
ax[4].imshow(im_data1)

plt.show()
```

```python
fig, ax = plt.subplots(5,1,figsize=(20,20)) # 3 color channels and 1 image plot

ax[0].hist(im_data2[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(im_data2[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(im_data2[:,:,2].flatten(), bins=256, color='blue') # Blue Channel
ax[3].hist(im_data2[:,:,3].flatten(), bins=256, color='orange') # Alpha Channel

ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')
ax[0].set_ylabel('Number of pixels')
ax[1].set_ylabel('Number of pixels')
ax[2].set_ylabel('Number of pixels')
ax[3].set_ylabel('Number of pixels')
# finally, include my image:
ax[4].imshow(im_data2)

plt.show()
```

```python
fig, ax = plt.subplots(5,1,figsize=(20,20)) # 4 color channels and 1 image plot

ax[0].hist(im_data3[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(im_data3[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(im_data3[:,:,2].flatten(), bins=256, color='blue') # Blue Channel
ax[3].hist(im_data3[:,:,3].flatten(), bins=256, color='orange') # Alpha Channel

ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')
ax[0].set_ylabel('Number of pixels')
ax[1].set_ylabel('Number of pixels')
ax[2].set_ylabel('Number of pixels')
ax[3].set_ylabel('Number of pixels')

# finally, include my image:
ax[4].imshow(im_data3)

plt.show()
```

## Observations:

### After doing the above analysis on the given 3 images we observe the following:
###  1. All the images consists of Red, green, blue and alpha channels. 
###  2. Image 1 majorly comprises of red channel and alpha channel. The blue and green channels are comparitively less. Moreover, it is quite evident from the image that the color Red is present in the image (when we look at it by eye), which is seen in histogram as well. 
###  3. Image 2 comprises of red , blue and green channel in almost equal proportions. The alpha channel bins are mostly seen on 0 and near 255.
###  4. Image 3 also depicts somewhat similar behavior as image 2. The red, blue and green channels are in equal proportions, whereas the alpha channel can be seen on 0 and near 255.
### 5. All the three images have major proportion of alpha channel (which depicts the level of opaqueness), which can be clearly seen when we plot the image and apply some interior color to it.(i applied red background as can be seen above)
### 6. Through these distributions we can clearly see a lot of things which have been mentioned above which cannot be merely seen by eye.
